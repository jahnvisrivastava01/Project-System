```python
import sqlite3

# Create database
conn = sqlite3.connect('employee_management.db')

# Cursor object
cursor = conn.cursor()

print("Database connected successfully!")
```

```python
cursor.execute(
    ''' CREATE TABLE Departments(
    DepartmentId INTEGER PRIMARY KEY AUTOINCREMENT,
    DepartmentName TEXT NOT NULL)
''')
conn.commit()
print("Departments table created!")
```

```python
cursor.execute("INSERT INTO Departments (DepartmentName) VALUES ('IT')")
cursor.execute("INSERT INTO Departments (DepartmentName) VALUES ('HR')")
cursor.execute("INSERT INTO Departments (DepartmentName) VALUES ('Finance')")
conn.commit()
```

```python
cursor.execute("SELECT * FROM Departments")
rows=cursor.fetchall()
for row in rows:
  print(row)
```

```python
cursor.execute('''
CREATE TABLE Employees(
    EmployeeId INTEGER PRIMARY KEY AUTOINCREMENT,
    Name TEXT NOT NULL,
    Email TEXT UNIQUE,
    HireDate DATE,
    DepartmentId INTEGER,
    FOREIGN KEY (DepartmentId)
    REFERENCES Departments(DepartmentId)
)
''')

conn.commit()

print("Employees table created successfully!")
```

```python
cursor.execute("""
SELECT e.Name, e.Email, d.DepartmentName
FROM Employees e
JOIN Departments d ON e.DepartmentId = d.DepartmentId
""")

for row in cursor.fetchall():
  print(row)
```

```python
cursor.execute('''
CREATE TABLE Projects(
    ProjectId INTEGER PRIMARY KEY AUTOINCREMENT,
    ProjectName TEXT NOT NULL,
    Technology TEXT,
    StartDate DATE
)
''')

conn.commit()

print("Projects table created successfully!")
```

```python
cursor.execute('''
CREATE TABLE EmployeeProjects(
    EmployeeId INTEGER,
    ProjectId INTEGER,
    AssignedDate DATE,

    PRIMARY KEY (EmployeeId, ProjectId),
    FOREIGN KEY(EmployeeId)
    REFERENCES Employees(EmployeeId) )
''')

conn.commit()

print("EmployeeProjects table created successfully!")
```

```python
cursor.execute("""
SELECT e.Name,
       p.ProjectName,
       ep.AssignedDate
FROM EmployeeProjects ep
JOIN Employees e
ON ep.EmployeeId = e.EmployeeId
JOIN Projects p
ON ep.ProjectId = p.ProjectId
""")

for row in cursor.fetchall():
    print(row)
```

```python
cursor.execute('''
CREATE TABLE Leaves(
LeaveId INTEGER PRIMARY KEY AUTOINCREMENT,
    EmployeeId INTEGER,
    LeaveType TEXT NOT NULL,
    StartDate DATE,
    EndDate DATE,
    Status TEXT,

    FOREIGN KEY (EmployeeId)
    REFERENCES Employees(EmployeeId)
)
''')

conn.commit()

print("Leaves table created successfully!")
```

```python
cursor.execute("""
SELECT e.Name,
       l.LeaveType,
       l.StartDate,
       l.EndDate,
       l.Status
FROM Employees e
JOIN Leaves l
ON e.EmployeeId = l.EmployeeId
""")

for row in cursor.fetchall():
    print(row)
```

```python
cursor.execute('''
CREATE TABLE Laptops(
    LaptopId INTEGER PRIMARY KEY AUTOINCREMENT,
    SerialNumber TEXT UNIQUE,
    Brand TEXT,
    EmployeeId INTEGER UNIQUE,

    FOREIGN KEY (EmployeeId)
    REFERENCES Employees(EmployeeId)
)
''')

conn.commit()

print("Laptops table created successfully!")
```
